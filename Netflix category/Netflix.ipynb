```python
import pandas as pd
import numpy as np
```

```python
netflix_data = pd.read_csv('netflix_titles.csv')
netflix_data.head()
```

```python
netflix_data.info()
```

```python
netflix_data['director'].fillna('Netflix', inplace=True)
```

```python
netflix_data['director'].value_counts()
```

```python
netflix_data.isna().sum()
```

```python
netflix_data.dropna(subset=['cast', 'country', 'date_added','rating'], inplace=True)
netflix_data.isna().sum()
len(netflix_data)
netflix_data
```

```python
# First changing the type into the numerical values 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

Type  = le.fit_transform(list(netflix_data['type']))
```

```python
X = netflix_data.drop(['type', 'director'], axis=1)
y = list(Type)
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ['title', 'cast' , 'country', 'date_added', 'rating', 'duration', 'listed_in', 'description']

one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', 
                                one_hot,
                                categorical_features)],
                               remainder="passthrough")

transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
```

```python
transformed_X
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100)

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
prediction = model.predict(X_test)
```
